INTRODUCTION

HEALTHCARE SECTOR IS CONSIDERED TO BE ONE OF THE NEXT BIG THINGS IN THE COMING DECADE. In THE WAKE OF COVID-19 PEOPLE HAVE BECOME FAR MORE CONSCIOUS ABOUT THEIR HEALTH AND WELLBEING THAN EVER BEFORE.THE WORLD HEALTH ORGANIZATION (WHO) RECOMMENDS A STANDARD OF 5 BEDS PER THOUSAND POPULATION. UNFORTUNATELY MANY OF THE STATES IN INDIA FALL SHORT OF THIS CUTOFF BY A CONSIDERABLE AMOUNT. SO MUCH SO THAT THE INDIAN GOVERNMENT HAD TO CONVERT RAILWAY COACHES INTO MOBILE HEALTH CARE UNITS IN ORDER TO HANDLE THE COVID-19 CRISIS . CERTAINLY OPENING ADEQUATE NUMBER OF HOSPITALS IN NICHE AREAS IS DEFINITELY AN OPTION THAT CAN'T BE OVERLOOKED WHETHER WE SEE IT FROM THE BUSINESS PERSPECTIVE OR FROM THE SOCIAL CAUSE PERSPECTIVE.IN THIS CAPSTONE PROJECT I AM GOING TO EXPLORE THE PROSPECTS OF OPENING HOSPITALS IN AREAS OF CHENNAI BY EXPLORING THE CURRENT SCENARIO AND THEN USING SEGMENTATION TO LOOK AT POTENTIAL CLUSTER REGIONS THAT MIGHT JUST PROVE TO BE AN ATTRACTIVE OPTION.

WITHOUT DOUBT CHENNAI IS ONE THE MOST POPULATED DISTRICT IN THE STATE OF TAMILNADU. AND SO IT AUTOMATICALLY QUALIFIES AS A MARKET WHICH CAN BE TARGETED UPON. AND OFCOURSE IN A METROPOLITAN CITY LIKE CHENNAI FINDING POTENTIAL DOCTORS AND OTHER PROFESSIONALS IS ALSO AN EASY JOB AS THE MAJORITY POPULATION IS WELL-EDUCATED.THE PROJECT WILL FIND OUT THE DISTRIBUTION OF HOSPITALS IN CHENNAI AND IDENTIFY POTENTIAL CLUSTERS WHERE THERE ARE LESS/NO HOSPITALS AND GIVES SUGGESTIONS FOR POTENTIAL LOCATIONS BASED ON THAT.
