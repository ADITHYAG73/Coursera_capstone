    AT FIRST I AM GOING TO LEVERAGE THE POWER OF FOURSQUARE DATA IN ORDER TO BE ABLE TO FIND THE DISTRIBUTION OF HOSPITALS IN AND AROUND NEIGHBORHOODS OF CHENNAI.ON AN ACCOUNT OF THIS I WILL ALSO BE RETRIEVING THE LIST OF HOSPITAL DATA FROM A PUBLICLY AVAILABLE URL AND ON A COMPARITIVE STUDY I WILL MAKE USE OF THE BEST AVAILABLE DATA IN ORDER TO PERFORM CLUSTERING ANALYSIS.

## RESULTS

FIRSTLY I EXPLORED THE POSSIBILITY OF IDENTIFYING HOSPITAL CLUSTERS BY SCRAPING THROUGH NEIGHBORHOOD DATA OF CHENNAI CITY.I SCRAPED THE WIKI PAGE OF CHENNAI CITY TO GET THE NEIGHBORHOODS.I THEN USED FOURSQUARE TO GET THE LIST OF HOSPITALS IN AND AROUND CHENNAI CITY.BUT NOT ENOUGH HOSPITALS COULD BE FOUND AND HENCE CLUSTERING COULD NOT BE MEANINGFUL.

NOW I EXTRACTED THE INFORMATION OF DIFFERENT HOSPITALS IN CHENNAI FROM ANOTHER WEBPAGE.NOW BY LEVERAGING THE FOURSQUARE DATA I CLUSTERED THE HOSPITALS INTO 5 CLUSTERS.

CLUSTER ANALYSIS OF THESE HOSPITALS WERE DISPLAYED ON A MAP.FROM THE MAP CONCLUSIONS WERE DRAWN WHICH WILL BE DISCUSSED IN THE CONCLUSION SECTION.

## CONCLUSIONS

FIRSTLY DENSE CLUSTERS OF HOSPITALS WERE OBSERVED IN THE COASTAL PART OF THE CITY AND DENSE CLUSTER WAS OBSERVED EVEN AT THE CENTRAL PART .BUT THE INTERIOR PARTS OF THE CITY WERE SPARSE WHEN IT COMES TO NUMBER OF HOSPITALS.AND ALSO THESE ARE THE LOW DEVELOPED REGIONS AS COMPARED TO OTHER PARTS OF THE CITY.THIS IS QUITE INTERESTING.ON THE OTHER HAND THE DECISION TO BUILD AN INFRASTRUCTURE OF VERY HIGH QUALITY IS UPTO THE STAKEHOLDERS AS THE PEOPLE OF THESE CLUSTERS MIGHT NOTBE ABLE TO AFFOR HIGH COST OF MEDICAL CARE THAT THESE HOSPITALS WILL PROVIDE.

## DISCUSSIONS

K-MEANS CLUSTERING TECHNIQUE THAT IS EMPLOYED IN THIS PROJECT IS LOW ACCURATE CLUSTERING MODEL AS COMPARED TO OTHER CLUSTERING TECHNIQUES LIKE DBSCAN.AND SO OTHER HIGH ACCURATE CLUSTERING ALGORITHMS MIGHT BE LOOKED UPON AS AN OPTION.
NOT TO FORGET THE PROJECT IDEA PROPOSED DID NOT TAKE INTO ACCOUNT THE DEMOGRAPHICS OF PEOPLE OF DIFFERENT NEIGHBORHOODS.ONE NEIGHBORHOOD MAY CONTAIN PEOPLE WHO MIGHT HAVE DIFFERENT INCOME LEVELS AS COMPARED TO OTHER NEIGHBORHOOD.THERFORE BUILDING HIGH QUALITY HOSPITALS MIGHT NOT DRAW THE ATTENTION OF POOR PEOPLE WHO CAN'T OBVIOUSLY AFFORD COSTLY MEDICAL CARE.AND SO DIFFERENT SOCIO-ECONOMIC FACTORS NEED TO BE CONSIDERED BEFORE ARRIVING AT A CONCLUSION WHETHER IT IS A VIABLE OPTION TO BUILD HOSPITALS IN AREAS WITH LESS NUMBER OF HOSPITALS AS RETURN OF INVESTMENT IS AN IMPORTANT PARAMETER FOR BUSINESSES IN LONG RUN.
